```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize = (5,2))
plt.scatter(x,y, c='r', alpha = 0.5, marker = '>')
plt.xlabel('This is My X -axis')
plt.ylabel('This is My Y -axis')
plt.title('This My Scatter Plot')
plt.grid()
plt.show()
```

```python
a = np.linspace(1,10,100)
b = np.sin(a)
```

```python
a
```

```python
b
```

```python
plt.scatter(a,b)
```

```python
plt.figure(figsize = (10,10))
plt.plot(a,b, '--r')
plt.xlabel('This is my X data')
plt.ylabel('This is my Y data')
plt.xlabel('This is sin wave graph')
plt.grid()
```

```python
i = ['a','b','c','d','e']
j=np.random.rand(5)
```

```python
plt.figure(figsize=(6,2))
plt.bar(i,j, color='g')
plt.xlabel('This is my Categorical data')
plt.ylabel('This is my Numerical data')
plt.xlabel('This is bar Plot')
plt.grid()
```

```python
plt.figure(figsize=(6,2))
plt.barh(i,j, color='g') #barh give this kind
plt.xlabel('This is my Categorical data')
plt.ylabel('This is my Numerical data')
plt.xlabel('This is bar Plot')
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```
