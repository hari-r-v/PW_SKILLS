### Decision Tree Classifier

- Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

- Loading dataset - iris

```python
data = load_iris()
```

```python
print(data.DESCR)
```

```python
df = sns.load_dataset('iris')
```

```python
data.target
```

```python
# split Independent and Dependent features

X = df.iloc[:,:-1]
y = data.target
```

- In decision Tree feature scaling/ standard scaling - because we gone a split nodes based on one particular feature, so it doesn't make any sense

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train , y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifer = DecisionTreeClassifier()
```

```python
classifer.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifer,filled = True)
```

```python
plt.figure(figsize=(12,10))
tree.plot_tree(classifer, filled= False)
```

- Observation :
  - the above split of tree based on gini impurity
  -  so lets create a model based on entropy

```python
classifer_entropy = DecisionTreeClassifier(criterion='entropy')
```

```python
classifer_entropy.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifer_entropy, filled= True)
```

- Observation :
    - Here we can make our decision after the 2 split only, but our model do split till 5 split which not necessary and it will leads to well trained classification model which may leads to overfitting problems
    -  so, we gone a do Pruning here

##### Post pruning

```python
classifer_entropy = DecisionTreeClassifier(criterion='entropy',max_depth=2)
classifer_entropy.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(12,10))
tree.plot_tree(classifer_entropy, filled= True)
```

```python
## Prediction
y_pred =classifer_entropy.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```

### What If we have a Huge data set???
    - In that case post-pruning we take lot of time conception, so in that case we will do pre-pruning

### Decision Tree PrePruning and Hyperparameter tuning for huge dataset

```python
import warnings
warnings.filterwarnings('ignore')
```

- Hyper parameter tuning

```python
parameter = {
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best','random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto','sqrt','log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifer=DecisionTreeClassifier()
```

```python
clf = GridSearchCV(classifer,param_grid=parameter, cv=5,scoring='accuracy')
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred,y_test)
print(score)
print(classification_report(y_pred,y_test))
```
